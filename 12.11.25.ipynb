```python
pop_std_dev=5.6
n=40
x_bar=32
```

```python
zcritical_80=stats.norm.ppf(0.90)
zcritical_80
```

```python
pop_mean_lower=x_bar-zcritical_80*(pop_std_dev/np.sqrt(n))
pop_mean_upper=x_bar+zcritical_80*(pop_std_dev/np.sqrt(n))
print(pop_mean_lower,pop_mean_upper)
```

```python
zcritical_90=stats.norm.ppf(0.95)
zcritical_90
```

```python
pop_mean_lower=x_bar-zcritical_90*(pop_std_dev/np.sqrt(n))
pop_mean_upper=x_bar+zcritical_90*(pop_std_dev/np.sqrt(n))
print(pop_mean_lower,pop_mean_upper)
```

```python
zcritical_98=stats.norm.ppf(0.99)
zcritical_98
```

```python
pop_mean_lower=x_bar-zcritical_98*(pop_std_dev/np.sqrt(n))
pop_mean_upper=x_bar+zcritical_98*(pop_std_dev/np.sqrt(n))
print(pop_mean_lower,pop_mean_upper)
```
