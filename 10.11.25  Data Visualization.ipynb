```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.colors as cl
print(cl.CSS4_COLORS)
```

```python
Exp=[1,2,3,4,5,6]
Sal=[1000,2500,3000,5000,8000,12000]
```

```python
print(plt.style.available)
```

```python
plt.style.use('default')
plt.scatter(Exp,Sal,color='skyblue',marker='>',linewidths=3)
plt.xlabel('Experience',size=14,color='Blue')   
plt.ylabel('Salary ',size=14,color='Red')
plt.title('Exp vs Sal',size=14,color='Green')
```

```python
plt.style.use('ggplot')
plt.scatter(Exp,Sal,color='blue',marker='>',linewidths=3)
plt.xlabel('Experience',size=14,color='Blue')   
plt.ylabel('Salary ',size=14,color='Red')
plt.title('Exp vs Sal',size=14,color='Green')
```

```python
### Linechart
plt.plot(Exp,Sal,marker='o',linestyle=':',markerfacecolor='red',markersize=8,lw=4)
```

```python
### BarChart
plt.bar(Exp,Sal,color='skyblue',edgecolor='black')
plt.ylim(0,16000)   # setting the Y axis limit 
```

```python
plt.bar_label(plt.bar(Exp,Sal,color='skyblue',edgecolor='black'))               # bar_label -> helps to mark the values of the plot 
plt.xlabel('Experience',size=14,color='Blue')   
plt.ylabel('Salary ',size=14,color='Red')
plt.title('Exp vs Sal',size=14,color='Green')
plt.ylim(0,16000)
```

```python
Marks=[23,34,56,67,78]
Names=['Ajay','Sanjay','John','Paul','Smith']
colors=['Blue','black','green','pink','wheat']
```

```python
plt.pie(Marks,labels=Names,colors=colors,textprops={'color':'orange','size':15},radius=1,
       wedgeprops={'linewidth':3,'edgecolor':'black'},autopct='%.f%%');
```

```python
### Subplots ->  to display diff plots in one plot
#(2,2)-> shows that its si 2x2 matrix type
#figsize=(x,y)-> height,width of plot
#axs -> takes the matrix values
```

```python
fig,axs=plt.subplots(2,2,figsize=(15,8))
axs[0,0].scatter(Exp,Sal,color='Brown',marker='*',linewidths=2)
axs[0,0].set_xlabel('Experience',color='Blue',size=10)   
axs[0,0].set_ylabel('Salary ',color='Red',size=10)
axs[0,0].set_title('Exp vs Sal',color='Green',size=10)
#subplot2
axs[0,1].plot(Exp,Sal,marker='o',linestyle=':',markerfacecolor='red',markersize=8,lw=4)
axs[0,1].set_xlabel('Experience',color='Blue',size=10)   
axs[0,1].set_ylabel('Salary ',color='Red',size=10)
axs[0,1].set_title('Exp vs Sal',color='Green',size=10)
axs[0,1].set_ylim(0,14000)
#subplot3
axs[1,0].bar(Exp,Sal,color='skyblue',edgecolor='black')
axs[1,0].set_xlabel('Experience',color='Blue',size=10)   
axs[1,0].set_ylabel('Salary ',color='Red',size=10)
axs[1,0].set_title('Exp vs Sal',color='Green',size=10)
axs[1,0].set_ylim(0,16000)
#subplot4
axs[1,1].pie(Marks,labels=Names,colors=colors,textprops={'color':'orange','size':15},radius=1,
       wedgeprops={'linewidth':3,'edgecolor':'black'},autopct='%.f%%',explode=[0,0,0,0.1,0]);
plt.tight_layout()
```
