```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.scatterplot(data=df,x='carat',y='price')
```

```python
### scatterplot with color of ur choice 
sns.scatterplot(data=df,x='carat',y='price',color='goldenrod')
```

```python
sns.histplot(data=df,x='carat')
```

```python
### kde 
sns.histplot(data=df,x='carat',color='olivedrab',kde='True')
```

```python
### density plot
sns.kdeplot(data=df,x='carat',fill=True,color='olivedrab',alpha=1)
```

```python
## Box plot  -> shows the outliers in the data
sns.boxplot(data=df,y='carat')
```

```python
## Violin plot-> combination of density plot+box plot
sns.violinplot(data=df,x='carat',color='olivedrab')
```

```python
sns.violinplot(data=df,y='carat',color='olivedrab')
```

```python
## countplot-> gives count of values of every column
df.cut.value_counts()
```

```python
sns.countplot(data=df,x='cut')
```

```python
sns.countplot(data=df,x='depth')
```

```python
### Univariant- one column
## Bivaraiant->  two column
##jointplot  -> gives both uni & bivariant in one plot
sns.jointplot(data=df,x='carat',y='price')
```

```python
### Pairplot -> o/p is in matrix form
### PP gives hist/density plot for same columns i.e diagonal columns
```

```python
sns.pairplot(data=df[['carat','depth','price','table']],diag_kind='kde')
```

```python
## without kde assigned histograms are plotted in diagonal columns, assigning  kde density plots r plotted in diagonal columns 
sns.pairplot(data=df[['carat','depth','price','table']])  
```

```python
## Correlation
S1=df[['carat','depth','price','table']].corr()
S1
```

```python
### Covariance
S2=df[['carat','depth','price','table']].cov()
S2
```

```python
df['carat'].skew()
```

```python
#### Heatmap is used to visualize when data is in matrix form
sns.heatmap(S1)
```

```python
sns.heatmap(S1,annot=True) 
```
